### 0.2 Indentation (The Golden Rule)
In many languages (C++, Java), you use `{ }` to group code. **In Python, we use Indentation (Spacing).**

If code is indented, it belongs to the block above it (like an `if` statement or a loop). Standard indentation is **4 Spaces** or **1 Tab**.

**Incorrect:**
```python
if True:
print("Error!") # No indentation
```

**Correct:**
```python
if True:
    print("Correct!") # Indented
```

````python
# Try running this to see proper indentation in action
if 10 > 5:
    print("10 is greater than 5")
    print("This line is also inside the 'if' block")

print("This line is OUTSIDE the block (it runs no matter what)")
````

## 🧩 Part 1: Python Core Fundamentals

### 1.1 Variables & Data Principles
Python is a **dynamically typed** language, meaning you don't need to declare the type of a variable explicitly. The interpreter infers it at runtime.

**Core Data Types:**
*   `int`: Whole numbers (e.g., 10, -5).
*   `float`: Decimal numbers (e.g., 3.14, 2.0).
*   `str`: Text data (e.g., "Hello").
*   `bool`: Logical values (`True` or `False`).

````python
# Variable Assignment
xp_points = 1500           # Integer
accuracy_rate = 0.985      # Float
user_name = "DataWizard123"   # String
is_admin = True            # Boolean

# Printing with Type Inspection
print(f"User: {user_name} (Type: {type(user_name)})")
print(f"XP: {xp_points} (Type: {type(xp_points)})")
print(f"Admin Access: {is_admin} (Type: {type(is_admin)})")
````

### 1.2 Data Structures (The Containers)
Data structures organize and store data. Choosing the right one is critical for performance.

| Structure | Syntax | Mutable? | Ordered? | Duplicates? | Description |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **List** | `[1, 2]` | ✅ Yes | ✅ Yes | ✅ Yes | General purpose array. |
| **Tuple** | `(1, 2)` | ❌ No | ✅ Yes | ✅ Yes | Immutable collection. Faster than lists. |
| **Set** | `{1, 2}` | ✅ Yes | ❌ No | ❌ No | Unique elements only. Fast lookups. |
| **Dict** | `{'k': 'v'}` | ✅ Yes | ✅ Yes | ❌ Keys | Key-Value mappings. The most powerful structure. |

````python
# --- LISTS (Mutable sequences) ---
skills = ["Python", "SQL", "Tableau"]
skills.append("Machine Learning")  # Adding an item
print(f"Skills List: {skills}")
print(f"First Skill: {skills[0]}")

# --- TUPLES (Read-only sequences) ---
db_config = ("192.168.1.5", 5432)  # (grapht, port)
# db_config[0] = "localhost"  # This would causing an ERROR
print(f"Database Port: {db_config[1]}")

# --- SETS (Unique items) ---
tags = {"AI", "ML", "AI", "Data", "ML"}
print(f"Unique Tags: {tags}")  # Duplicates automatically removed

# --- DICTIONARIES (Key-Value) ---
model_metrics = {
    "accuracy": 0.89,
    "f1_score": 0.87,
    "recall": 0.92
}
print(f"Model Accuracy: {model_metrics['accuracy']}")
````

### 1.3 Control Flow logic
Programming is about making decisions (`if-else`) and repeating tasks (`loops`).

#### 🌟 The Star Pattern Exercise
A classic logic building exercise. We use nested loops to print patterns.

````python
print("--- Standard Loop ---")
rows = 5
for i in range(1, rows + 1):
    # Inner loop isn't needed here if we use string multiplication
    # '*' * 3 results in '***'
    print("*" * i)

print("\n--- List Comprehension (Advanced) ---")
# Creating a list of squares for even numbers only
numbers = range(10)
print(list(numbers))
even_squares = [n**2 for n in numbers if n % 2 == 0]
print(f"Even Squares: {even_squares}")
````

## ⚙️ Part 2: Advanced Python Concepts

### 2.1 Exception Handling
Code crashes. Good developers handle crashes gracefully. We use `try`, `except`, and `finally` blocks.

*   **Try**: Run this code.
*   **Except**: If an error happens, run this.
*   **Finally**: Always run this (good for cleanup).

````python
def safe_division(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero!"
    except TypeError:
        return "Error: Inputs must be numbers!"
    finally:
        print("Execution attempt finished.")

print(safe_division(10, 2))
print(safe_division(10, 0))
print(safe_division(10, "two"))
````

### 2.2 Functional Programming: Lambda, Map, Filter
*   **Lambda**: Anonymous, one-line functions.
*   **Map**: Apply a function to every item in a list.
*   **Filter**: Keep only items that satisfy a condition.

````python
nums = [1, 2, 3, 4, 5]

# Square every number using Map + Lambda
squared = list(map(lambda x: x**2, nums))
print(f"Squared: {squared}")
# Keep only numbers > 2 using Filter + Lambda
big_nums = list(filter(lambda x: x > 2, nums))
print(f"Numbers > 2: {big_nums}")
````

## 📂 Part 3: File I/O Management

Handling external data is crucial. Python allows reading and writing in various modes.

| Mode | Description |
| :--- | :--- |
| `'r'` | **Read** (Default). Error if file doesn't exist. |
| `'w'` | **Write**. Overwrites entire file. Creates if not exists. |
| `'a'` | **Append**. Adds to end of file. |
| `'r+'`| Read and Write. |

````python
import os
import json

# 1. Writing a Text File
with open("logs.txt", "w") as f:
    f.write("System Start: 10:00 AM\n")
    f.write("User Login: Admin\n")

# 2. Appending to a Text File
with open("logs.txt", "a") as f:
    f.write("Action: Update Database\n")

print("--- Reading Log File ---")
with open("logs.txt", "r") as f:
    print(f.read())

# 3. Handling JSON Data
config = {
    "theme": "dark",
    "notifications": True,
    "max_retries": 5
}

with open("config.json", "w") as f:
    json.dump(config, f, indent=4)

print("\n--- Loaded JSON Config ---")
with open("config.json", "r") as f:
    loaded = json.load(f)
    print(loaded)
````

## 🔢 Part 4: NumPy (Numerical Python)

<img src="https://upload.wikimedia.org/wikipedia/commons/3/31/NumPy_logo_2020.svg" width="200"/>

NumPy is the foundation of scientific computing. It replaces Python's slow lists with fast **Arrays**.

````python
import numpy as np

# Creating Arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Vectorized Operations (So much faster than loops!)
print(f"Addition: {a + b}")
print(f"Multiplication: {a * b}")

# Matrix Operations
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.eye(2) # Identity matrix

print(f"\nDot Product:\n{np.dot(matrix_a, matrix_b)}")

# Reshaping
grid = np.arange(9).reshape(3, 3)
print(f"\n3x3 Grid:\n{grid}")
````

## 🐼 Part 5: Pandas (Data Analysis)

<img src="https://upload.wikimedia.org/wikipedia/commons/e/ed/Pandas_logo.svg" width="200"/>

Pandas is the 'Excel' of Python. We work with **DataFrames**.

### 5.1 Creating and Merging Data

````python
import pandas as pd

# Dataset A: Employees
df_emp = pd.DataFrame({
    'ID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Dept_ID': ['D1', 'D2', 'D1']
})
print(df_emp)
# Dataset B: Departments
df_dept = pd.DataFrame({
    'Dept_ID': ['D1', 'D2', 'D3'],
    'Dept_Name': ['HR', 'Tech', 'Finance']
})
print(df_dept)
# Merging (SQL Join)
merged = pd.merge(df_emp, df_dept, on='Dept_ID', how='left')
print("--- Merged DataFrame ---")
print(merged)
````

### 5.2 Time Series Analysis
Pandas is incredibly strong at handling time-stamped data.

````python
# Generate date range
dates = pd.date_range(start='2023-01-01', periods=6, freq='D')
df_time = pd.DataFrame({
    'Date': dates,
    'Sales': [100, 120, 90, 110, 130, 150]
})

# Set Date as Index
df_time.set_index('Date', inplace=True)

# Resampling (e.g., Calculate 2-Day sum)
resampled = df_time.resample('2D').sum()
print("--- Resampled (2-Day Sum) ---")
print(resampled)
````

### 6.2 Word Clouds
Visualizing text frequency (Great for NLP).

````python
from wordcloud import WordCloud

text_data = "Python is amazing. Python is fast. Data Science requires Python. Machine Learning uses Python."

wc = WordCloud(width=800, height=400, background_color='black', colormap='rainbow').generate(text_data)

plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.title("Word Frequency Cloud")
plt.show()
````

### 6.3 Geospatial Maps (Folium)
Interactive coding maps.

````python
import folium

# Center View on India
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5, tiles="OpenStreetMap")

# Add Markers
locations = [
    {"loc": [28.6139, 77.2090], "name": "New Delhi", "info": "Capital"},
    {"loc": [19.0760, 72.8777], "name": "Mumbai", "info": "Finance Hub"},
    {"loc": [12.9716, 77.5946], "name": "Bangalore", "info": "Tech Hub"}
]

for place in locations:
    folium.Marker(
        location=place["loc"],
        popup=f"<b>{place['name']}</b><br>{place['info']}",
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(m)

m
````
